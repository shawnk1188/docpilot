```python
import torch
import transformers
import numpy as np

from sentence_transformers import SentenceTransformer

print(f"Torch version: {torch.__version__}")
print(f"Transformers version: {transformers.__version__}")

# This should now work without the "Disabling PyTorch" error
model = SentenceTransformer('all-MiniLM-L6-v2')

texts = [
    "The patient showed elevated blood pressure",
    "The doctor noted hypertension in the exam",    # semantically similar
    "The stock market dropped 3% on Friday",         # unrelated
]

vecs = model.encode(texts, normalize_embeddings=True)

# cosine similarity (dot product of normalised vectors)
print("similar pair:", np.dot(vecs[0], vecs[1]).round(3))   # expect > 0.7
print("unrelated pair:", np.dot(vecs[0], vecs[2]).round(3)) # expect < 0.3
```
